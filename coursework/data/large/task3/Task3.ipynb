```python
import pandas as pd 
```

```python
df = pd.read_csv('../titles.tsv', sep='\t', header=None, names="tconst,titleType,primaryTitle,originalTitle,isAdult,startYear,endYear,runtimeMinutes,genres".split(","))
df
```

```python
df = df[(df["startYear"] != "\\N") & (df["titleType"] != "\\N") & (df["genres"] != "\\N") & (df["primaryTitle"]!="\\N")]
df
```

```python
df['startYear'] = df['startYear'].astype(int)
```

```python
df = df[(1900 <= df["startYear"]) & (df["startYear"] <= 1999) & (df["titleType"] == "movie")]
df
```

```python
dfr = pd.read_csv('../title.ratings.tsv', sep='\t', header=0, dtype={"rating": float, "votes": int})
dfr
```

```python
dfm = pd.merge(dfr, df,  how='left', on="tconst")
dfm
```

```python
dfm = dfm.dropna()
dfm
```

```python
dfm['genres'] = dfm['genres'].str.split(',')
dfm
```

```python
dfm = dfm.explode("genres")
dfm
```

```python
dfm["startYear"] = dfm["startYear"].astype('int').astype("str").str[2] # convert to decade
dfm
```

```python
dfm = dfm.sort_values("primaryTitle")
dfm
```

```python
dfm['averageRating'] = dfm['averageRating'].astype(float)
dfm
```

```python
# dfm.groupby("startYear").agg({'averageRating': ['max'], "primaryTitle" : []}).reset_index()
largest = dfm.groupby("startYear")["averageRating"].nlargest(1)
# largest
dfm.iloc[[i[1] for i in largest.index], :].head(100)
```

```python
max_ratings = dfm.groupby(["startYear", "genres"]).agg({"averageRating": ["max"]}).reset_index()
max_ratings.columns = ["decade", "genres", "max"]
max_ratings
```

```python
with open("task3.csv", "w", encoding="utf-8") as out:
    for index, row in max_ratings.iterrows():
        for _, result in dfm[(dfm["startYear"] == row["decade"]) & (dfm["averageRating"] == row["max"]) & (dfm["genres"] == row["genres"])].sort_values("primaryTitle").iterrows():
            res = "%s|%s|%s" % (row["decade"], row["genres"], result.primaryTitle)
            print(res)
            out.write(res+"\n")
            break
```
