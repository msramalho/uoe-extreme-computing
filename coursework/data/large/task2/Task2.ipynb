```python
import pandas as pd
```

```python
df = pd.read_csv('../titles.tsv', sep='\t', header=None, names="tconst,titleType,primaryTitle,originalTitle,isAdult,startYear,endYear,runtimeMinutes,genres".split(","))
df
```

```python
# filter bad duration and genres
df = df[(df["startYear"] != "\\N") & (df["primaryTitle"]!="\\N")]
df
```

```python
df['startYear'] = df['startYear'].apply(pd.to_numeric)
```

```python
dfr = pd.read_csv('../title.ratings.tsv', sep='\t', header=0, dtype={"rating": float, "votes": int})
dfr
```

```python
df = df[(1990 <= df["startYear"]) & (df["startYear"] <= 2018) & (df["titleType"] == "movie")]
df
```

```python
dfr = dfr[(dfr["averageRating"]>=7.5)  & (dfr["numVotes"]>=50_000)]
dfr
```

```python
res = pd.merge(dfr, df,  how='left', on="tconst")
res
```

```python
res = res.dropna()
res
```

```python
films = sorted(res.primaryTitle.tolist())
```

```python
with open("task2.csv", "w", encoding="utf-8") as out:
    for f in films:
        out.write(f+"\n")
```

```python
with open("res_task2.csv", "r", encoding="utf-8") as inf:
    res_films = sorted([l.strip() for l in  inf.readlines()])
```
