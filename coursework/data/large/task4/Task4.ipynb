```python
import pandas as pd
```

```python
dfn = pd.read_csv('../name.basics.tsv', sep='\t', header=0)
dfn
```

```python
dfn = dfn[(dfn["primaryName"] != "\\N") & (dfn["knownForTitles"] != "\\N")]
dfn
```

```python
dfn['knownForTitles'] = dfn['knownForTitles'].str.split(',')
dfn = dfn.explode("knownForTitles")
dfn
```

```python
dfc = pd.read_csv('../title.crew.tsv', sep='\t', header=0)
dfc
```

```python
dfc = dfc[(dfc["writers"] != "\\N")]
dfc
```

```python
dfc['writers'] = dfc['writers'].str.split(',')
dfc = dfc.explode("writers")
dfc
```

```python
dfr = pd.read_csv('../title.ratings.tsv', sep='\t', header=0, dtype={"rating": float, "votes": int})
dfr
```

```python
df1 = pd.merge(dfr, dfc,  how='left', on="tconst")
df1
```

```python
df1 = df1.dropna()
df1
```

```python
df2 = pd.merge(df1, dfn,  how='left', left_on=["tconst", "writers"], right_on=["knownForTitles", "nconst"])
df2 = df2.dropna()
df2
```

```python
df2.sort_values("numVotes", ascending=False).head(10)
```

```python
with open("task4.csv", "w", encoding="utf-8") as out:
    for w in df2.sort_values("numVotes", ascending=False).head(10)["primaryName"].tolist():
        print(w)
        out.write(w+"\n")
```
