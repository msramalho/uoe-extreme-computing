```python
import pandas as pd
```

```python
df = pd.read_csv('../titles.tsv', sep='\t', header=0)
df.head()
```

```python
# filter bad duration and genres
df = df[(df["runtimeMinutes"] != "\\N") & (df["genres"]!="\\N")]
df.head()
```

```python
df['runtimeMinutes'] = df['runtimeMinutes'].apply(pd.to_numeric)
```

```python
df['genres'] = df['genres'].str.split(',')
df.head()
```

```python
df = df.explode("genres")
df.head()
```

```python
t1 = df.groupby("genres").agg({'runtimeMinutes': ['min', 'max', 'mean']}).reset_index()
t1.columns = [col[-1] for col in t1.columns.values]
t1 = t1[t1.columns[[3,2,1,0]]]
t1.rename(columns={ t1.columns[-1]: "genre" }, inplace = True)
t1 = t1.sort_values(by=['genre'])
t1["mean"] = t1["mean"].round(2)
t1
```

```python
t1.to_csv("task1.csv", header=None, index=False, sep="|")
```
